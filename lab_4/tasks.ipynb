```python
import requests
from bs4 import BeautifulSoup

headers = {
    'authority': 'www.google.com',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'accept-language': 'en-US,en;q=0.9',
    'cache-control': 'max-age=0',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36',
}
## number of films, will be scraped
size = 30
response = requests.get("https://www.imdb.com/chart/top", headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
##movies = soup.find_all("h3", class_="ipc-title__text")
##movies = [str(i) for i in movies]
##movies = [i[28:] for i in movies][1:]
##movies = [i[:-5] for i in movies][:250]
##print(movies)

movie_containers = soup.find_all('li', class_='ipc-metadata-list-summary-item sc-3f724978-0 enKyEL cli-parent')[:size]

titles = []
years = []
descriptions = []

for container in movie_containers:

    text = container.find_all('h3',class_='ipc-title__text')
    titles.append(text[0].text)
##<span class="sc-43986a27-8 jHYIIK cli-title-metadata-item">1972</span>
    year = container.find_all('span', class_='sc-43986a27-8 jHYIIK cli-title-metadata-item')
    years.append(year[0].text)
##<a href="/title/tt0068646/?ref_=chttp_t_2" class="ipc-title-link-wrapper" tabindex="0"><h3 class="ipc-title__text">2. The Godfather</h3></a>
    ref = str(container.find('a', class_="ipc-title-link-wrapper"))[40:]
    ref = ref[:32]
    resp = requests.get('https://www.imdb.com'+ref, headers=headers)
##<p data-testid="plot" class="sc-466bb6c-3 fOUpWp"></p>
    ref_soup = BeautifulSoup(resp.text, "html.parser")
    ref_container = ref_soup.find('p', class_="sc-466bb6c-3 fOUpWp")
##<span role="presentation" data-testid="plot-xl" class="sc-466bb6c-2 chnFO"></span>
##<span role="presentation" data-testid="plot-xs_to_m" class="sc-466bb6c-0 hlbAws"></span>
    desc = ref_container.find_all('span', class_='sc-466bb6c-0 hlbAws')
    descriptions.append(desc[0].text)

print(titles)
print(years)
print(descriptions)

```
