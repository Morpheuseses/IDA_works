```python
import matplotlib.pyplot as plt
import requests
import json
```

```python
json_link = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')

data = json_link.json()

print(data)
```

```python
dates = []

dates.append(data['Date'])

for _ in range(40):
    json_link = requests.get('https:' + data['PreviousURL'])
    data = json_link.json()
    dates.append(data['Date'])
```

```python
currencies = {}

for x in data["Valute"]:
    currencies[x] = []

values = currencies

json_link = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
data = json_link.json()

for i in range(40):
    valute_dict = data["Valute"].items()
    for k, x in valute_dict:
        values[k].append(x["Value"])
    json_link = requests.get('https:'+data["PreviousURL"])
    data = json_link.json()
```
